```python
import pandas as pd
deathrates = pd.read_csv("../data/raw/deathrates.csv")
max_death_row = deathrates.loc[deathrates['ESTIMATE'].idxmax()]
max_death_row

```

```python
import matplotlib.pyplot as plt
```

```python
# First, filter out the rows where 'STUB_NAME' is 'Sex' as indicated in the user's code snippet
#filtered_data_without_sex = deathrates[deathrates['STUB_NAME'] != 'Sex']
#pd.set_option('display.max_rows', 500)

# Now, we repeat the previous steps but with the filtered data
grouped_data = deathrates.groupby(['STUB_NAME', 'YEAR']).agg({'ESTIMATE': 'mean'}).reset_index()
pivot_data = grouped_data.pivot(index='YEAR', columns='STUB_NAME', values='ESTIMATE')
display(pivot_data)

plt.figure(figsize=(15, 8))
for race in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[race], label=race)

plt.title('Death Rates for Suicide by Race Over Years')
plt.xlabel('Year')
plt.ylabel('Death Rate per 100,000')
plt.legend(title='Race')
plt.grid(True)
plt.show()
```
